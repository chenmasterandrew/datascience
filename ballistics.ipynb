```python
from math import radians, sin, pi
from numpy import linspace
import matplotlib.pyplot as plt
```

```python
def rangeformula(v, theta):
    """
    Inputs:
    v - velocity in m/s
    theta - launch angle with respect to horizon in radians
    Returns: 
    Impact distance in m
    """
    return (v**2 / 9.81 * sin(2*theta))
```

```python
angledegrees = linspace(0,90,num=200)
```

```python
rangefromdegrees = []
for a in angledegrees:
    rangefromdegrees.append(rangeformula(100,radians(a)))
```

```python
plt.plot(angledegrees,rangefromdegrees)
plt.title('Range as a function of angle at 100 m/s')
plt.xlabel('Angle (degrees)')
plt.ylabel('Range (meters)')
plt.show()
```

```python
rangeforangle100 = [rangeformula(100, radians(angle)) for angle in angledegrees]
rangeforangle200 = [rangeformula(200, radians(angle)) for angle in angledegrees]
plt.plot(angledegrees, rangeforangle100, label="100 m/s")
plt.legend()
plt.plot(angledegrees, rangeforangle200, label="200 m/s")
plt.legend()
plt.title('Range as a function of angle at various velocities')
plt.xlabel('Angle (degrees)')
plt.ylabel('Range (meters)')
plt.show()
```

```python
angleradians = [radians(a) for a in angledegrees]
rangefromradians = [rangeformula(100,a) for a in angleradians]
plt.plot(angleradians,rangefromradians)
plt.title('Range as a function of angle at 100 m/s')
plt.xlabel('Angle (radians)')
plt.ylabel('Range (meters)')
plt.show()
```

```python
velocities = linspace(0,100,num=200)
rangefromvelocities45 = [rangeformula(v,radians(45)) for v in velocities]
plt.plot(velocities,rangefromvelocities45)
plt.title('Range as a function of velocity at 45 degrees')
plt.xlabel('Velocity (meters per second)')
plt.ylabel('Range (meters)')
plt.show()
```

```python
rangefromvelocities15 = [rangeformula(v,radians(15)) for v in velocities]
rangefromvelocities30 = [rangeformula(v,radians(30)) for v in velocities]
rangefromvelocities60 = [rangeformula(v,radians(60)) for v in velocities]
rangefromvelocities75 = [rangeformula(v,radians(75)) for v in velocities]
plt.plot(velocities, rangefromvelocities15, label="15 degrees")
plt.legend()
plt.plot(velocities, rangefromvelocities30, label="30 degrees")
plt.legend()
plt.plot(velocities, rangefromvelocities45, label="45 degrees")
plt.legend()
plt.plot(velocities, rangefromvelocities60, label="60 degrees")
plt.legend()
plt.plot(velocities, rangefromvelocities75, label="75 degrees")
plt.legend()
plt.title('Range as a function of velocity at various angles')
plt.xlabel('Velocity (meters per second)')
plt.ylabel('Range (meters)')
plt.show()
```
